```python
ml_boundary = shapefile.Reader(os.path.join(shapefile_pth, 'model_boundary.shp')) # Model boundary shapefile
mlb_shape = np.array(np.rint(ml_boundary.shapeRecords()[0].shape.points)) # Model boundary array
#orbacken = shapefile.Reader(os.path.join(shapefile_pth, 'orbacken_sfr.shp')) # Creek Örbäcken shapefile
#orbacken_shp = np.concatenate([np.rint(i.shape.points) for i in orbacken.shapeRecords()]) # Creek Örbäcken array
mlb_shapely = Polygon(mlb_shape)
#orb_shapely = LineString(orbacken_shp)
mlb_shapely_prep = prep(mlb_shapely)
```

```python
xmin, xmax, ymin, ymax = 426900, 427700, 6654650, 6655350 # These are the same coordinates used to construct the base-grid
resolution_sto = 33 # Equal space (in meters) between pilot points
pps_sto_initial = []
for lat in np.arange(xmin, xmax, resolution_sto):
    for lon in np.arange(ymin, ymax, resolution_sto):
        pps_sto_initial.append(Point((round(lat,4), round(lon,4))))
```

```python
pps_sto_pos = [] # All pps within model domain
for point in pps_sto_initial:
    if mlb_shapely_prep.contains(point):
        pps_sto_pos.append(point)
print(f'Number of points: {len(pps_sto_pos)}')
```

```python
def replace_in_file(path, input_file, original_str, replacement_str):
    # Read in the file
    with open(os.path.join(path, input_file), 'r') as file :
        filedata = file.read()

    # Replace the target string
    filedata = filedata.replace(original_str, replacement_str)

    # Write the file out again
    with open(os.path.join(path, input_file), 'w') as file:
        file.write(filedata)
```

Create (and test) a dedicated model for adjusting rtp parameter values. The parameters to be adjusted are `window` (1 - 100, initially 20) and `offset` (0.01 - 1.5, initially 0.5):

def model_rtp(window, offset):
    import math
    return rtp_model_input_data.rolling(window=int(math.ceil(window)),center=True,min_periods=1).mean() + offset

```python
def expand_diff( arr, delta ):
    diff = np.diff( arr, prepend = 0.0 )
    diff[1:] += np.sign( diff[1:] ) * delta
    new_arr = diff.cumsum()
    return new_arr + arr.mean() - new_arr.mean()
```

```python
#Create (and test) a dedicated model for adjusting rtp parameter values.
#The parameters to be adjusted are `expansion_coefficient` (0.001 - 0.2, initially 0.1) and `offset` (0.01 - 1.5, initially 0.5):
def model_rtp(dataset, expansion_coefficient, sigma, offset):
    
    temp_rtp = gaussian_filter1d(dataset, sigma) # Apply gaussian filter
    
    temp_rtp = expand_diff(temp_rtp, expansion_coefficient) # Apply expansion coefficient
    
    new_rtp = temp_rtp + offset
    
    return new_rtp
```

```python
def apply_ghb_multiplier():
    df = ghb_perdata_df
    
    # Get number of cells in for each of the bounds that we want to parameterize
    len_yellow = len(df.loc[df['bhead'] == 'ghb_yellow']) # Number of cells in yellow GHB cells
    len_blue = len(df.loc[df['bhead'] == 'ghb_blue']) # Number of cells in yellow GHB cells
    len_green = len(df.loc[df['bhead'] == 'ghb_green']) # Number of cells in yellow GHB cells
    
    # PARAMETERS - Suggested parval1: 0.01, parubnd: 0.9, parlbnd:0.001
    ghbyellomult, ghbbluemult, ghbgreenmult = 0.01, 0.01, 0.001
    
    # Linear interpolation. Suggested parval1: 0.01, parubnd: 0.9, parlbnd:0.001
    yellow_multiplier = np.linspace(1, ghbyellomult, len_yellow)
    blue_multiplier = np.linspace(1, ghbbluemult, len_blue)
    green_multiplier = np.linspace(1, ghbgreenmult, len_green)[::-1] # Flip it
    
    # Change data
    df.loc[df['bhead'] == 'ghb_yellow', 'multiplier'] = yellow_multiplier
    df.loc[df['bhead'] == 'ghb_blue', 'multiplier'] = blue_multiplier
    df.loc[df['bhead'] == 'ghb_green', 'multiplier'] = green_multiplier
    
    return df
```

## Parameterize LUMPREM models

Parameters to be adjusted through history-matching are:


| parameter | default | initial | upper bound | lower bound |
| --------- |--------:|--------:|-------------| ------------|
| maxvol    | 0.5     | 0.2     | 1           | 0.01        |
| rdelay    | 5       | 5       | 10          | 1           |
| mdelay    | 1       | 1       | 5           | 0           |
| Ks        | 0.1     | 86.4    | 8640.0      | 0.0000864   |
| m         | 0.5     | 1.0     | 2.0         | 0.01        |
| l         | 0.5     | 0.5     | 1.0         | 0.0         |
| mflowmax  | 0.1     | 0.1     | 0.2         | 0.01        |
| f         | 0.5     | 0.5     | 0.9         | 0.01        |
| γ         | 1.0     | 1.0     | 10.0        | 0.01        |

- *maxvol* The volume of the soil moisture store.
- *rdelay* Delay, in days, between water draining from the soil moisture store.
- *mdelay* Delay, in days, between water leaving the soil moisture store.
- *Ks* Hydraulic conductivity (meters per day). Initial: 86.4 (0.001 m/s)
- *m* A parameter determining the shape of the drainage rate vs. stored water relationship.
- *l* A pore-connectivity parameter (estimated by Mualem, 1976, to be about 0.5 for many soils).
- *mflowmax* Maximum macropore recharge allowed per day.
- *f* Crop factor.
- *γ* A parameter determining the shape of the evaporation rate vs. stored water relationship.

```python
def parameterize_lr_rch_model(lr_template_file, lr_ml_name):
    # Read in the file
    with open(os.path.join(template_ws, lr_template_file), 'r') as file:
        lines = file.readlines()

    # Replace the target string
    lines[4] = f'$ks_{lr_ml_name}$  $m_{lr_ml_name}$  0.5  0.1 \n' #earth properties factors
    lines[40] = f'$crfac_{lr_ml_name}$  $gamma_{lr_ml_name}$  \n' #crop factor & gamma
    
    # Write the file out again
    with open(os.path.join(template_ws, lr_template_file), 'w') as file:
        for line in lines:
            file.write(line)
```

Parameterize recharge model:

```python
parameterize_lr_rch_model('lr_lu1.tpl', 'lu1')
```

Parameterize GHB models:

```python
for file in os.listdir(template_ws):
    if 'lr_' in file and 'lu1' not in file:
        parameterize_lr_rch_model(file, file[3:][:3])
```

```python
def parameterize_lr_ghb_model(lr_template_file, lr_ml_name):
    # Read in the file
    with open(os.path.join(template_ws, lr_template_file), 'r') as file:
        lines = file.readlines()

    # Replace the target string
    lines[6] = f'$of_{lr_ml_name}$  $f1_{lr_ml_name}$  $f2_{lr_ml_name}$  $p_{lr_ml_name}$  -9999.0 10000.0\n' #earth properties factors
    
    # Write the file out again
    with open(os.path.join(template_ws, lr_template_file), 'w') as file:
        for line in lines:
            file.write(line)
```
